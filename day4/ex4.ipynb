```python
# change image to black and white image using numpy

import numpy as np
import imageio
import matplotlib.pyplot as plt

pic = imageio.imread('demo_2.jpg')

#plt.figure(figsize=(16,12))
#plt.imshow(pic)
#plt.show()
print(type(pic))
print(pic.shape)

print("Height:",pic.shape[0])
print("Width:",pic.shape[1])
```

```python
g = lambda img: np.dot(img[...,:3],[0.299,0.587,0.114])

gray = g(pic)
plt.figure(figsize=(16,12))
plt.imshow(gray, cmap=plt.get_cmap(name='gray'))
plt.show()
```
