```python
# numpy examples
import numpy as np

a = [5,8,12,15]

arr1 = np.array(a)

print(type(arr1))
print(arr1)

b = [17,5,21,6]

arr2 = np.array(b)

print(arr2)

c = a+b

arr3 = arr1*arr2

print(c)
print(arr3)
```

```python
# shape

arr1 = np.array([5,8,12,15])

print(arr1.shape)

arr1 = np.array (
    [
        [5,8,12,15],
        [17,5,21,6]
    ]
)

print(arr1)
print(arr1.shape)

arr1 = np.array (
    [
        [
        [5,8,12,15],
        [17,5,21,6]
        ],
        [
        [5,8,12,15],
        [17,5,21,6]
        ],
        [
        [5,8,12,15],
        [17,5,21,6]
        ]
    ]
    
)

print(arr1.shape)
```

```python
#[x1,x2] * [y1,y2]  = (x1*y1+x2*y2)

arr1 = np.array([12,24]).reshape(-1,1)

print(arr1)
print(arr1.shape)
```

```python
# linspace - generate values between a small range

arr1 = np.linspace(1,10,200)

print(arr1)
```
